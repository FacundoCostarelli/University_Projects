## **WORK INSTRUCTION #2**
Implement a Python script that performs the following actions:

* Generate a vector "$t$" between 0 and $4\pi$ (seconds) with time steps of $T_s=0.001$
* Calculate $sin(2t)$ and store it in the vector $Sa$
* Calculate $cos(4t)$ and store it in the vector $Sb$
* Generate a function called CALC that provides the sum and element-wise product between $Sa$ and $Sb$
* Plot in the same window: one axis with $Sa$ in blue, another with $Sb$ in red, a third with the sum (dotted green), and the fourth with the product result (dashed magenta). The axes should be adjusted, with corresponding labels and a legend identifying each functional relationship.

Explanatory comments should be placed for each statement.

**HELP:** *Use the Python reference manual (numpy and matplotlib) for the use of the arange, sin, cos, and plot functions.*

### **[ESP] BLOQUE 1 - CONSIGNA 2**
### **[ENG] CELL 1 - WORK INSTRUCTION 2**

```python
# ESP: Módulo NUMPY (Permite realizar operaciones matemáticas)
# ENG: NUMPY Module (Allows performing mathematical operations)
import numpy as np

# ESP: Módulo MATPLOTLIB (Permite implementar visualización gráfica)
# ENG: MATPLOTLIB Module (Allows implementing graphical visualization)
import matplotlib.pyplot as plt

# ESP: Módulo SCIPY (Permite realizar el procesamiento de señales)
# ENG: SCIPY Module (Allows performing signal processing)
import scipy.signal as sp
```

### **[ESP] BLOQUE 2 - CONSIGNA 2**
### **[ENG] CELL 2 - WORK INSTRUCTION 2**

```python
# ESP: CÓDIGO SOLUCIÓN
# ENG: SOLUTION CODE
# AUTOR: FACUNDO COSTARELLI

# ESP: Función CALC
# ENG: CALC Function
# ESP: Toma dos arrays como entrada y devuelve la suma y producto punto a punto, es decir, elemento a elemento.
# ENG: Takes two arrays as input and returns the element-wise sum and product.
def CALC(arr1, arr2):
  """
  ESP: Calcula la suma y el producto elemento a elemento de dos arrays.
  ENG: Calculates the element-wise sum and product of two arrays.

  Args:
    arr1: El primer array.
    arr2: El segundo array.

  Returns:
    Una tupla que contiene dos arrays: (suma_de_arrays, producto_de_arrays).
    A tuple containing two arrays: (sum_of_arrays, product_of_arrays).
  """
  # ESP: Suma elemento a elemento
  # ENG: Element-wise sum
  suma = arr1 + arr2
  # ESP: Producto elemento a elemento
  # ENG: Element-wise product
  producto = arr1 * arr2
  return suma, producto
```

```python
# ESP: DESARROLLO DE LA CONSIGNA - Ejemplo INICIAL
# ENG: INSTRUCTION DEVELOPMENT - Initial Example
Ts=0.001
t=np.arange(0,4*np.pi,Ts)
Sa=np.sin(2*t)
Sb=np.cos(4*t)
plt.subplot(311)
plt.plot(t,Sa,'b')
plt.subplot(312)
plt.plot(t,Sb,'r')
```

```python
# ESP: CÓDIGO SOLUCIÓN
# ENG: SOLUTION CODE
# AUTOR: FACUNDO COSTARELLI

# ESP: Invoco a la función CALC
# ENG: Invoking the CALC function
suma_Sa_Sb, producto_Sa_Sb = CALC(Sa, Sb)

# ESP: Imprimo las dimensiones o cantidad de elementos de los arrays para verificar
# ENG: Print the dimensions or number of elements of the arrays to verify
print(f"Dimensiones de t: {t.shape}")
print(f"Dimensiones de Sa: {Sa.shape}")
print(f"Dimensiones de Sb: {Sb.shape}")

# ESP: Imprimo las dimensiones o cantidad de elementos de las variables de resultados para verificar
# ENG: Print the dimensions or number of elements of the result variables to verify
print(f"Dimensiones de la suma: {suma_Sa_Sb.shape}")
print(f"Dimensiones del producto: {producto_Sa_Sb.shape}")
```

```python
# ESP: Configuracion de la figura, grafica, ajuste de ejes, etiquetas y leyenda
# ENG: Figure, graph, axis adjustment, labels, and legend configuration
# ESP: Crear una figura con 4 subplots en una columna
# ENG: Create a figure with 4 subplots in one column
fig, axes = plt.subplots(4, 1, figsize=(10, 12), sharex=True) # sharex=True para compartir el eje x

# ESP: Graficar Sa en el primer subplot (azul)
# ENG: Plot Sa in the first subplot (blue)
axes[0].plot(t, Sa, 'b', label='Sa = sen(2t)')
axes[0].set_xlabel('Tiempo (segundos)')
axes[0].set_ylabel('Amplitud')
axes[0].set_title('Señal Sa')
axes[0].grid(True)
axes[0].legend()

# ESP: Graficar Sb en el segundo subplot (rojo)
# ENG: Plot Sb in the second subplot (red)
axes[1].plot(t, Sb, 'r', label='Sb = cos(4t)')
axes[1].set_xlabel('Tiempo (segundos)')
axes[1].set_ylabel('Amplitud')
axes[1].set_title('Señal Sb')
axes[1].grid(True)
axes[1].legend()

# ESP: Graficar la suma en el tercer subplot (verde punteado)
# ENG: Plot the sum in the third subplot (dotted green)
axes[2].plot(t, suma_Sa_Sb, 'g-.', label='Suma (Sa + Sb)')
axes[2].set_xlabel('Tiempo (segundos)')
axes[2].set_ylabel('Amplitud')
axes[2].set_title('Señal Sa+Sb')
axes[2].grid(True)
axes[2].legend()

# ESP: Graficar el producto en el cuarto subplot (magenta a rayas)
# ENG: Plot the product in the fourth subplot (dashed magenta)
axes[3].plot(t, producto_Sa_Sb, 'm--', label='Producto (Sa * Sb)')
axes[3].set_xlabel('Tiempo (segundos)')
axes[3].set_ylabel('Amplitud')
axes[3].set_title('Señal Sa*Sb')
axes[3].grid(True)
axes[3].legend()

# ESP: Ajustar el diseño para evitar solapamiento
# ENG: Adjust layout to prevent overlapping
plt.tight_layout()

# ESP: Mostrar la gráfica
# ENG: Show the graph
plt.show()
```
